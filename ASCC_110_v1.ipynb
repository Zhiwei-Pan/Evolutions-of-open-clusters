```python
import numpy as np
from astropy.io import fits 
import matplotlib.pyplot as plt
import astropy.constants as c
import astropy.units as u
import scipy.constants as spc
import pandas as pd
from astropy.table import QTable


D_fits=fits.open('ASCC_110_0.3.fits')
data=D_fits[1].data

plx=data['parallax']
plx_error=data['parallax_error']
astroerror=data['astrometric_excess_noise']
pmra=data['pmra']
pmdec=data['pmdec']
pmra_error=data['pmra_error']
pmdec_error=data['pmdec_error']
ra=data['ra']
ra_error=data['ra_error']
dec=data['dec']
dec_error=data['dec_error']

#index=(astroerror<1)&(np.abs(pmra_error/pmra)<0.1)&(np.abs(pmdec_error/pmdec)<0.1)&(ra_error<1/6)&(dec_error<1/6)&(np.abs(plx_error/plx)<0.1)
index=(astroerror<1)&(abs(plx_error/plx)<0.1)
data=data[index]

pmra_error=data['pmra_error']
pmdec_error=data['pmdec_error']
ra=data['ra']
dec=data['dec']
pmra=data['pmra']
pmdec=data['pmdec']
plx=data['parallax']

print(len(plx))
```

```python
plx_plot_max=2
plx_plot_min=0
plx_plot_d=0.1

plx_plot_bin=np.arange(plx_plot_min,plx_plot_max,plx_plot_d)

font=20
for each in range(len(plx_plot_bin)):
    index=(plx>0.9*plx_plot_bin[each])&(plx<1.1*plx_plot_bin[each])
    data1=data[index]
    
    pmdec=data1['pmdec']
    plt.figure(figsize=(8,6))
    str1='pmra distribution'+str(0.7*plx_plot_bin[each])+'-'+str(1.3*plx_plot_bin[each])
    n, bins, patches = plt.hist(x=pmdec, bins='auto',alpha=0.7, rwidth=0.85)
    plt.title(str1,fontsize=font)
    plt.xlabel('pmdec',fontsize=font)
    plt.ylabel('Number',fontsize=font)
    plt.tick_params(labelsize=font)
    plt.show()
```
